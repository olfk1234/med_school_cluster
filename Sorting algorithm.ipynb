```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.environ["OMP_NUM_THREADS"] = '1'
```

```python
df = pd.read_excel("med_school_stats.xlsx")
```

```python
df
```

```python
df.columns = ["State", "School", "Num Applications", "In State Application %", "Out of State %", "Men Application %", 
              "Women Application %", "Num Matriculants", "In State Matriculant %", 
              "Out of State Matriculant %", "Men Matriculant %", "Women Matriculant %", "Average GPA", "Average MCAT"]
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
X = df.copy()
X.columns
```

```python
X = X.drop(["State", "School", 'Num Applications', 'In State Application %', 
        'Out of State %', 'Men Application %', 'Women Application %', 'Num Matriculants', "Men Matriculant %",'Women Matriculant %', 'Out of State Matriculant %'], axis = 1)
```

```python
X = X.reset_index().drop("index", axis = 1)
```

```python
X
```

```python
from sklearn.metrics import silhouette_score
```

```python
X.columns
```

```python
X = X.astype("float")
```

```python
X
```

CLUSTERING ALGORITMS WITHOUT PCA

```python
sns.pairplot(X)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

KMEANS WITHOUT PCA

```python
val_list = []
silhouette_score_list = []
for n in range(2,100):
    kmean = KMeans(n_clusters=n, random_state=101, n_init = "auto")
    labels = kmean.fit_predict(X_sc)
    val_list.append(kmean.inertia_)
    silhouette_score_list.append(silhouette_score(X_sc,kmean.labels_))
```

```python
# pip install -U threadpoolctl --user
```

```python
plt.plot(range(2,100), val_list, marker = "o")
```

```python
plt.plot(range(2,100), silhouette_score_list, marker = "o")
```

```python
kmean = KMeans(n_clusters=4, random_state=101, n_init = "auto")
labels = kmean.fit_predict(X_sc)
```

```python
X_two = X.copy()
X_two["label"] = kmean.labels_
sns.pairplot(X_two, hue = "label", palette="Set1")
```

```python
silhouette_score(X_sc, labels)
```

K-means WITH PCA -- The model used in Web Deployment

```python
from sklearn.decomposition import PCA
vr = []
for n in range(1,4):
    model = PCA(n_components=n, random_state=101)
    model.fit(X_sc)
    vr.append(sum(model.explained_variance_ratio_))
plt.plot(range(1,4), vr)
```

```python
model = PCA(n_components=2, random_state=101)
X_sc_PCA = model.fit_transform(X_sc)
```

```python
X_sc_PCA
```

```python
df_PCA = pd.DataFrame(X_sc_PCA, columns = ["Feature 1", "Feature 2"])
```

```python
df_PCA
```

```python
val_list = []
silhouette_score_list = []
for n in range(2,100):
    kmean = KMeans(n_clusters=n, random_state=101, n_init = "auto")
    labels = kmean.fit_predict(X_sc_PCA)
    val_list.append(kmean.inertia_)
    silhouette_score_list.append(silhouette_score(X_sc_PCA,kmean.labels_))
plt.plot(range(2,100), val_list, marker = "o")
```

```python
plt.plot(range(2,100), silhouette_score_list, marker = "o")
```
